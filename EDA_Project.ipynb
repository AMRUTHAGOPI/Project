```python
numeric_cols = ['Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours',
                'BMI', 'Heart_Rate','Physical_Activity_Hours']

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Gender', y='Sleep_Hours')
plt.title('Sleep Hours by Gender')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
sns.boxplot(
    data=data,
    x='Gender',
    y='Sleep_Hours',
    hue='Alcohol_Consumption',
    palette='Set2'  # You can change this to any palette you like
)
plt.title('Sleep Hours by Gender and Alcohol Consumption')
plt.xlabel('Gender')
plt.ylabel('Sleep Hours')
plt.legend(title='Alcohol Consumption', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=data[['Age', 'Coffee_Intake', 'Caffeine_mg', 'BMI', 'Heart_Rate', 'Gender']], hue='Gender')
plt.show()
```

```python
country_caffeine = data.groupby('Country')['Caffeine_mg'].mean().sort_values(ascending=False)
country_caffeine.plot(kind='bar', figsize=(10, 5), title='Average Caffeine consumption(in mg) by Country', color='green')
plt.xlabel('Country')
plt.ylabel('Average Caffeine(mg)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by country and calculate means
country_stats = data.groupby('Country').agg({
    'Caffeine_mg': 'mean',
    'Sleep_Hours': 'mean',
}).reset_index()

# Plot Caffeine vs Sleep Hours with country labels
plt.figure(figsize=(12, 6))
plt.scatter(country_stats['Caffeine_mg'], country_stats['Sleep_Hours'])

# Add country names as labels
for i, row in country_stats.iterrows():
    plt.text(row['Caffeine_mg'], row['Sleep_Hours'], row['Country'], fontsize=9, alpha=0.75)

plt.title('Average Caffeine Intake vs Sleep Hours by Country')
plt.xlabel('Average Caffeine (mg)')
plt.ylabel('Average Sleep Hours')
plt.grid(True)
plt.show()




```

```python
import seaborn as sns

plt.figure(figsize=(14, 7))
gender_country_caffeine = data.groupby(['Country', 'Gender'])['Caffeine_mg'].mean().reset_index()

sns.barplot(data=gender_country_caffeine, x='Country', y='Caffeine_mg', hue='Gender')
plt.title('Average Caffeine Consumption by Gender and Country')
plt.ylabel('Average Caffeine (mg)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate sample size and average caffeine per country
country_summary = data.groupby('Country').agg(
    Avg_Caffeine=('Caffeine_mg', 'mean'),
    Sample_Size=('Caffeine_mg', 'count')
).reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(data=country_summary, x='Country', y='Avg_Caffeine', palette='viridis')
plt.title('Average Caffeine Consumption by Country with Sample Size')
plt.ylabel('Average Caffeine (mg)')
plt.xticks(rotation=45)

# Add sample size text labels on top of bars
for i, row in country_summary.iterrows():
    plt.text(i, row['Avg_Caffeine'] + 0.5, f"N={row['Sample_Size']}", ha='center')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Sleep_Quality', y='Sleep_Hours', palette='Set3')
plt.title('Distribution of Sleep Hours by Sleep Quality')
plt.xlabel('Sleep Quality')
plt.ylabel('Sleep Hours')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=data, x='Stress_Level', y='Heart_Rate', palette='coolwarm')
plt.title('Swarm Plot of Heart Rate by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Heart Rate')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns
numeric_cols = ['Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours',
                'BMI', 'Heart_Rate', 'Physical_Activity_Hours']

# Calculate correlation matrix
corr = data[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```
